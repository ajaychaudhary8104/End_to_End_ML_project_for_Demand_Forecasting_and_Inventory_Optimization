# Demand Forecasting and Inventory Optimization

## 1. Problem Definition

### 1.1 Business Background

Retail organizations operate in highly dynamic environments where customer demand fluctuates due to seasonality, promotions, economic conditions, weather patterns, product characteristics, and regional purchasing behavior. Accurately predicting future demand is critical for maintaining optimal inventory levels and ensuring product availability across stores.

Traditional inventory planning methods often rely on historical averages or manual forecasting processes, which fail to capture complex demand patterns and external influencing factors. As a result, retailers frequently experience:

* Stockouts leading to lost sales and dissatisfied customers.
* Overstock situations causing increased holding and warehousing costs.
* Inefficient procurement and replenishment decisions.
* Poor utilization of working capital.
* Reduced profitability and operational efficiency.

To address these challenges, organizations are increasingly adopting Machine Learning and MLOps-driven demand forecasting systems that leverage historical sales data and external business signals to generate accurate demand predictions.

---

## 1.2 Business Problem Statement

The retail company needs an intelligent forecasting system capable of predicting future product demand at the individual store level.

Without accurate demand forecasts, the organization faces:

### Stockout Risk

* Sales opportunities are lost.

* Brand reputation may be affected.

### Overstock Risk

When inventory exceeds actual demand:

* Storage costs increase.

### Supply Chain Inefficiencies

The company therefore requires a data-driven forecasting solution capable of improving demand visibility and supporting inventory optimization decisions.

## 1.9 Success Criteria

### Technical Success Metrics

#### Mean Absolute Error (MAE)

Measures average prediction error.

[
MAE = \frac{1}{n}\sum |y_i - \hat{y_i}|
]

Lower values indicate better performance.

---

#### Root Mean Squared Error (RMSE)

Penalizes large prediction errors.

[
RMSE = \sqrt{\frac{1}{n}\sum(y_i-\hat{y_i})^2}
]

---

#### Mean Absolute Percentage Error (MAPE)

Measures percentage forecasting accuracy.

[
MAPE = \frac{100}{n}
\sum
\left|
\frac{y_i-\hat{y_i}}{y_i}
\right|
]

Target:

* MAPE < 15%

Stretch Goal:

* MAPE < 10%

````python
from dataclasses import dataclass

@dataclass
class GeneratorConfig:

    start_date: str = "2022-01-01"
    end_date: str = "2023-12-31"

    n_products: int = 100
    n_stores: int = 10
    n_suppliers: int = 10

    random_state: int = 42

    output_dir: str = "data/raw"
````

````python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from faker import Faker
fake = Faker()


class DemandInventoryDataGenerator:

    def __init__(self, config):

        self.config = config

        np.random.seed(
            config.random_state
        )

        self.dates = pd.date_range(
            config.start_date,
            config.end_date,
            freq="D"
        )

        os.makedirs(
            config.output_dir,
            exist_ok=True
        )

    # =====================================================
    # SUPPLIER MASTER
    # =====================================================

    def generate_suppliers(self):

        suppliers = pd.DataFrame({

            "supplier_id":
            range(
                1,
                self.config.n_suppliers + 1
            ),

            "supplier_rating":
            np.random.uniform(
                3.5,
                5.0,
                self.config.n_suppliers
            ),

            "avg_lead_time":
            np.random.randint(
                3,
                20,
                self.config.n_suppliers
            ),

            "lead_time_std":
            np.random.randint(
                1,
                5,
                self.config.n_suppliers
            )
        })

        return suppliers

    def generate_products(self):

        categories = [
            "Electronics",
            "Groceries",
            "Fashion",
            "Home",
            "Sports"
        ]

        products = pd.DataFrame({

            "product_id":
            range(
                1,
                self.config.n_products + 1
            ),

            "category":
            np.random.choice(
                categories,
                self.config.n_products
            ),

            "supplier_id":
            np.random.randint(
                1,
                self.config.n_suppliers + 1,
                self.config.n_products
            ),

            "unit_cost":
            np.random.uniform(
                10,
                500,
                self.config.n_products
            ),

            "margin_pct":
            np.random.uniform(
                0.2,
                0.5,
                self.config.n_products
            )
        })

        products["unit_price"] = (
            products["unit_cost"]
            *
            (1 + products["margin_pct"])
        )

        return products

    def generate_stores(self):

        stores = pd.DataFrame({

            "store_id":
            range(
                1,
                self.config.n_stores + 1
            ),

            "region":
            np.random.choice(
                [
                    "North",
                    "South",
                    "East",
                    "West"
                ],
                self.config.n_stores
            ),

            "store_size_sqft":
            np.random.randint(
                1000,
                15000,
                self.config.n_stores
            ),

            "store_type":
            np.random.choice(
                [
                    "Hypermarket",
                    "Supermarket",
                    "Convenience"
                ],
                self.config.n_stores
            )
        })

        return stores

    def generate_calendar(self):

        calendar = pd.DataFrame({

            "date":
            self.dates
        })

        calendar["year"] = (
            calendar["date"].dt.year
        )

        calendar["month"] = (
            calendar["date"].dt.month
        )

        calendar["week"] = (
            calendar["date"]
            .dt.isocalendar()
            .week
            .astype(int)
        )

        calendar["dayofweek"] = (
            calendar["date"].dt.dayofweek
        )

        calendar["is_weekend"] = (
            calendar["dayofweek"] >= 5
        ).astype(int)

        return calendar

    def generate_weather(self):

        calendar = pd.DataFrame({
            "date": self.dates
        })

        day_of_year = (
            calendar["date"]
            .dt.dayofyear
        )

        weather = pd.DataFrame({

            "date":
            self.dates,

            "temperature":
            (
                25
                +
                10*np.sin(
                    2*np.pi*
                    day_of_year/365
                )
                +
                np.random.normal(
                    0,
                    2,
                    len(self.dates)
                )
            ),

            "rainfall":
            np.random.gamma(
                2,
                3,
                len(self.dates)
            ),

            "humidity":
            np.random.uniform(
                40,
                90,
                len(self.dates)
            )
        })

        return weather

    def generate_economics(self):

        return pd.DataFrame({

            "date":
            self.dates,

            "inflation_rate":
            np.random.normal(
                5,
                0.3,
                len(self.dates)
            ),

            "fuel_price":
            np.random.normal(
                100,
                10,
                len(self.dates)
            ),

            "consumer_index":
            np.random.normal(
                100,
                5,
                len(self.dates)
            )
        })

    def generate_promotions(self):

        n_promotions = 50000

        return pd.DataFrame({

            "date":
            np.random.choice(
                self.dates,
                n_promotions
            ),

            "product_id":
            np.random.randint(
                1,
                self.config.n_products+1,
                n_promotions
            ),

            "store_id":
            np.random.randint(
                1,
                self.config.n_stores+1,
                n_promotions
            ),

            "discount_pct":
            np.random.choice(
                [5,10,15,20,25,30],
                n_promotions
            )
        })

    def generate_sales(self, products, stores, calendar, weather):

        sales = (
            calendar.assign(key=1).merge(
                products.assign(key=1),
                on="key"
            ).merge(
                stores.assign(key=1),
                on="key"
            ).drop(
                "key",
                axis=1
            )
        )

        sales = sales.merge(
            weather,
            on="date",
            how="left"
        )

        category_multiplier = {

            "Electronics": 1.5,
            "Groceries": 3.0,
            "Fashion": 1.8,
            "Home": 1.2,
            "Sports": 1.4
        }

        sales["base_demand"] = (

            sales["category"].map(
                category_multiplier
            ) * 20
        )

        seasonality = (  1
            +
            0.2*np.sin(
                2*np.pi*
                sales["month"]/12
            )
        )

        demand = (

            sales["base_demand"]
            *
            seasonality
            *
            (
                1
                +
                sales["is_weekend"]*0.15
            )
            *
            (
                1
                +
                (
                    sales["temperature"]-25
                )*0.01
            )
        )

        sales["units_sold"] = (
            np.random.poisson(
                np.maximum(
                    demand,
                    1
                )
            )
        )

        sales["revenue"] = (
            sales["units_sold"]
            *
            sales["unit_price"]
        )

        return sales

    def generate_inventory(self,sales):

        inventory = sales[[
            "date",
            "product_id",
            "store_id"
        ]].copy()

        inventory["stock_on_hand"] = np.random.randint(
            100,
            2000,
            len(inventory)
        )

        inventory["stock_in_transit"] = np.random.randint(
            0,
            500,
            len(inventory)
        )

        inventory["backorder_qty"] = np.random.randint(
            0,
            20,
            len(inventory)
        )

        return inventory

    def run(self):

        suppliers = self.generate_suppliers()

        products = self.generate_products()

        stores = self.generate_stores()

        calendar = self.generate_calendar()

        weather = self.generate_weather()

        economics = self.generate_economics()

        promotions = self.generate_promotions()

        sales = self.generate_sales(
            products,
            stores,
            calendar,
            weather
        )

        inventory = self.generate_inventory(
            sales
        )

        suppliers.to_parquet(
            "data/raw/suppliers.parquet"
        )

        products.to_parquet(
            "data/raw/products.parquet"
        )

        stores.to_parquet(
            "data/raw/stores.parquet"
        )

        weather.to_parquet(
            "data/raw/weather.parquet"
        )

        economics.to_parquet(
            "data/raw/economics.parquet"
        )

        promotions.to_parquet(
            "data/raw/promotions.parquet"
        )

        sales.to_parquet(
            "data/raw/sales.parquet"
        )

        inventory.to_parquet(
            "data/raw/inventory.parquet"
        )

        print("Data generation complete.")
````

## 7.DEMAND PATTERN ANALYSIS

````python
# ============================================================
# STEP 7: DEMAND PATTERN ANALYSIS
# ============================================================
# Objectives:
# 1. Product Demand Trends
# 2. Store Demand Trends
# 3. Category Demand Trends
# 4. Regional Demand Trends
# 5. Weekly Seasonality
# 6. Monthly Seasonality
# 7. Weekend vs Weekday Demand
# 8. Promotion Impact Analysis
# 9. Weather Impact Analysis
# 10. Economic Impact Analysis
# 11. Demand Volatility Analysis
# 12. ABC Analysis
# 13. XYZ Analysis
# 14. Product Segmentation (ABC-XYZ Matrix)
# 15. Business Insights
# ============================================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

print("=" * 120)
print("STEP 7: DEMAND PATTERN ANALYSIS")
print("=" * 120)

df = master_df.copy()

TARGET = "units_sold"

# ============================================================
# BASIC INFORMATION
# ============================================================

print("\nDataset Shape:", df.shape)

if "date" in df.columns:
    print("Date Range:", df["date"].min(), "to", df["date"].max())

# ============================================================
# 1. OVERALL DEMAND TREND
# ============================================================

print("\nAnalyzing Overall Demand Trend...")

daily_demand = (
    df.groupby("date")[TARGET]
    .sum()
    .reset_index()
)

plt.figure(figsize=(15,5))
plt.plot(daily_demand["date"], daily_demand[TARGET])
plt.title("Daily Demand Trend")
plt.xlabel("Date")
plt.ylabel("Units Sold")
plt.grid(True)
plt.show()

# ============================================================
# 2. PRODUCT DEMAND ANALYSIS
# ============================================================

print("\nTop Products by Demand")

product_demand = (
    df.groupby("product_id")[TARGET]
    .agg(["sum","mean","std"])
    .sort_values("sum", ascending=False)
)

display(product_demand.head(10))

plt.figure(figsize=(12,5))
product_demand["sum"].head(20).plot(kind="bar")
plt.title("Top 20 Products by Demand")
plt.ylabel("Units Sold")
plt.show()

# ============================================================
# 3. STORE DEMAND ANALYSIS
# ============================================================

if "store_id" in df.columns:

    print("\nStore Performance")

    store_demand = (
        df.groupby("store_id")[TARGET]
        .agg(["sum","mean"])
        .sort_values("sum", ascending=False)
    )

    display(store_demand.head(10))

    plt.figure(figsize=(12,5))
    store_demand["sum"].head(20).plot(kind="bar")
    plt.title("Top Stores by Demand")
    plt.ylabel("Units Sold")
    plt.show()

# ============================================================
# 4. CATEGORY DEMAND ANALYSIS
# ============================================================

if "category" in df.columns:

    print("\nCategory Performance")

    category_demand = (
        df.groupby("category")[TARGET]
        .agg(["sum","mean"])
        .sort_values("sum", ascending=False)
    )

    display(category_demand)

    plt.figure(figsize=(10,5))
    category_demand["sum"].plot(kind="bar")
    plt.title("Demand by Category")
    plt.ylabel("Units Sold")
    plt.show()

# ============================================================
# 5. REGION DEMAND ANALYSIS
# ============================================================

if "region" in df.columns:

    print("\nRegional Demand Analysis")

    region_demand = (
        df.groupby("region")[TARGET]
        .agg(["sum","mean"])
        .sort_values("sum", ascending=False)
    )

    display(region_demand)

    plt.figure(figsize=(8,5))
    region_demand["sum"].plot(kind="bar")
    plt.title("Demand by Region")
    plt.ylabel("Units Sold")
    plt.show()

# ============================================================
# 6. WEEKLY SEASONALITY
# ============================================================

if "dayofweek" in df.columns:

    print("\nWeekly Seasonality")

    weekly_pattern = (
        df.groupby("dayofweek")[TARGET]
        .mean()
    )

    display(weekly_pattern)

    plt.figure(figsize=(8,5))
    weekly_pattern.plot(marker="o")
    plt.title("Average Demand by Day of Week")
    plt.ylabel("Units Sold")
    plt.grid(True)
    plt.show()

# ============================================================
# 7. MONTHLY SEASONALITY
# ============================================================

if "month" in df.columns:

    print("\nMonthly Seasonality")

    monthly_pattern = (
        df.groupby("month")[TARGET]
        .mean()
    )

    display(monthly_pattern)

    plt.figure(figsize=(10,5))
    monthly_pattern.plot(marker="o")
    plt.title("Average Demand by Month")
    plt.ylabel("Units Sold")
    plt.grid(True)
    plt.show()

# ============================================================
# 8. WEEKEND VS WEEKDAY
# ============================================================

if "is_weekend" in df.columns:

    print("\nWeekend vs Weekday Demand")

    weekend_analysis = (
        df.groupby("is_weekend")[TARGET]
        .agg(["mean","sum"])
    )

    weekend_analysis.index = ["Weekday","Weekend"]

    display(weekend_analysis)

    plt.figure(figsize=(6,4))
    weekend_analysis["mean"].plot(kind="bar")
    plt.title("Weekend vs Weekday Demand")
    plt.ylabel("Average Units Sold")
    plt.show()

# ============================================================
# 9. PROMOTION IMPACT
# ============================================================

if "discount_pct" in df.columns:

    print("\nPromotion Impact Analysis")

    df["promotion_flag"] = np.where(
        df["discount_pct"] > 0,
        "Promotion",
        "No Promotion"
    )

    promo_analysis = (
        df.groupby("promotion_flag")[TARGET]
        .agg(["mean","sum","count"])
    )

    display(promo_analysis)

    plt.figure(figsize=(6,4))
    promo_analysis["mean"].plot(kind="bar")
    plt.title("Promotion Impact on Demand")
    plt.ylabel("Average Units Sold")
    plt.show()

# ============================================================
# 10. WEATHER IMPACT
# ============================================================

weather_features = [
    "temperature",
    "rainfall",
    "humidity"
]

existing_weather = [
    col for col in weather_features
    if col in df.columns
]

if existing_weather:

    print("\nWeather Correlation with Demand")

    weather_corr = (
        df[existing_weather + [TARGET]]
        .corr()[TARGET]
        .sort_values(ascending=False)
    )

    display(weather_corr)

# ============================================================
# 11. ECONOMIC IMPACT
# ============================================================

economic_features = [
    "inflation_rate",
    "fuel_price",
    "consumer_index"
]

existing_economic = [
    col for col in economic_features
    if col in df.columns
]

if existing_economic:

    print("\nEconomic Indicators Correlation")

    econ_corr = (
        df[existing_economic + [TARGET]]
        .corr()[TARGET]
        .sort_values(ascending=False)
    )

    display(econ_corr)

# ============================================================
# 12. DEMAND VOLATILITY
# ============================================================

print("\nDemand Volatility Analysis")

volatility = (
    df.groupby("product_id")[TARGET]
    .agg(["mean","std"])
)

volatility["cv"] = (
    volatility["std"] /
    volatility["mean"]
)

display(
    volatility.sort_values(
        "cv",
        ascending=False
    ).head(10)
)

# ============================================================
# 13. ABC ANALYSIS
# ============================================================

print("\nPerforming ABC Analysis")

abc = (
    df.groupby("product_id")[TARGET]
    .sum()
    .reset_index()
)

abc = abc.sort_values(
    TARGET,
    ascending=False
)

abc["cum_sales"] = abc[TARGET].cumsum()

abc["cum_pct"] = (
    abc["cum_sales"] /
    abc[TARGET].sum()
) * 100

abc["ABC_Class"] = np.where(
    abc["cum_pct"] <= 80,
    "A",
    np.where(
        abc["cum_pct"] <= 95,
        "B",
        "C"
    )
)

display(
    abc["ABC_Class"]
    .value_counts()
)

# ============================================================
# 14. XYZ ANALYSIS
# ============================================================

print("\nPerforming XYZ Analysis")

xyz = (
    df.groupby("product_id")[TARGET]
    .agg(["mean","std"])
)

xyz["cv"] = (
    xyz["std"] /
    xyz["mean"]
)

xyz["XYZ_Class"] = np.where(
    xyz["cv"] <= 0.25,
    "X",
    np.where(
        xyz["cv"] <= 0.50,
        "Y",
        "Z"
    )
)

display(
    xyz["XYZ_Class"]
    .value_counts()
)

# ============================================================
# 15. ABC-XYZ MATRIX
# ============================================================

abc_xyz = (
    abc[["product_id","ABC_Class"]]
    .merge(
        xyz[["XYZ_Class"]],
        left_on="product_id",
        right_index=True
    )
)

abc_xyz["Segment"] = (
    abc_xyz["ABC_Class"] +
    abc_xyz["XYZ_Class"]
)

print("\nABC-XYZ Matrix")

segment_table = (
    abc_xyz["Segment"]
    .value_counts()
    .sort_index()
)

display(segment_table)

# ============================================================
# 16. DEMAND DISTRIBUTION
# ============================================================

plt.figure(figsize=(10,5))
sns.histplot(
    df[TARGET],
    bins=50,
    kde=True
)
plt.title("Demand Distribution")
plt.show()

# ============================================================
# 17. BUSINESS INSIGHTS
# ============================================================

print("\n" + "="*120)
print("BUSINESS INSIGHTS")
print("="*120)

print(f"""
1. Total Records Analysed: {len(df):,}

2. Average Daily Demand:
   {daily_demand[TARGET].mean():.2f}

3. Highest Demand Product:
   {product_demand.index[0]}

4. Highest Demand Store:
   {store_demand.index[0] if 'store_id' in df.columns else 'N/A'}

5. Most Valuable ABC Segment:
   A-Class Products

6. Most Stable Demand Segment:
   X-Class Products

7. High Priority Inventory Segment:
   AX Products

8. High Risk Segment:
   CZ Products

9. Products requiring Safety Stock:
   Y and Z Products

10. Recommended Forecast Granularity:
    Product × Store × Date
""")

print("\nDemand Pattern Analysis Completed Successfully.")
print("=" * 120)
````

````python
# ============================================================
# STEP 8: INVENTORY ANALYSIS
# ============================================================
# Objectives:
# 1. Inventory Overview
# 2. Stock-On-Hand Analysis
# 3. Inventory Distribution
# 4. Inventory by Product
# 5. Inventory by Store
# 6. Inventory by Category
# 7. Inventory by Region
# 8. Inventory Turnover Analysis
# 9. Days of Inventory (DOI)
# 10. Stockout Risk Analysis
# 11. Backorder Analysis
# 12. Inventory Coverage Analysis
# 13. Supplier Lead Time Analysis
# 14. ABC Inventory Classification
# 15. Inventory Cost Analysis
# 16. Inventory Health Score
# 17. Business Insights & Recommendations
# ============================================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

print("=" * 120)
print("STEP 8 : INVENTORY ANALYSIS")
print("=" * 120)

df = master_df.copy()

# ============================================================
# REQUIRED COLUMNS CHECK
# ============================================================

required_cols = [
    "units_sold",
    "stock_on_hand"
]

missing_cols = [
    col for col in required_cols
    if col not in df.columns
]

if missing_cols:
    raise ValueError(f"Missing required columns: {missing_cols}")

# ============================================================
# 1. INVENTORY OVERVIEW
# ============================================================

print("\nINVENTORY OVERVIEW")

inventory_summary = pd.DataFrame({
    "Metric": [
        "Total Records",
        "Total Products",
        "Total Stores",
        "Avg Stock On Hand",
        "Median Stock On Hand",
        "Max Stock On Hand",
        "Min Stock On Hand"
    ],
    "Value": [
        len(df),
        df["product_id"].nunique(),
        df["store_id"].nunique(),
        round(df["stock_on_hand"].mean(),2),
        round(df["stock_on_hand"].median(),2),
        round(df["stock_on_hand"].max(),2),
        round(df["stock_on_hand"].min(),2)
    ]
})

display(inventory_summary)

# ============================================================
# 2. STOCK ON HAND DISTRIBUTION
# ============================================================

plt.figure(figsize=(12,5))
sns.histplot(
    df["stock_on_hand"],
    bins=50,
    kde=True
)
plt.title("Stock On Hand Distribution")
plt.xlabel("Inventory Units")
plt.show()

# ============================================================
# 3. INVENTORY STATISTICS
# ============================================================

print("\nSTOCK ON HAND STATISTICS")

display(
    df["stock_on_hand"]
    .describe()
    .to_frame("Stock On Hand")
)

# ============================================================
# 4. PRODUCT LEVEL INVENTORY
# ============================================================

print("\nTOP PRODUCTS BY INVENTORY")

product_inventory = (
    df.groupby("product_id")
    .agg({
        "stock_on_hand":"mean",
        "units_sold":"mean"
    })
    .sort_values(
        "stock_on_hand",
        ascending=False
    )
)

display(product_inventory.head(20))

plt.figure(figsize=(14,5))
product_inventory["stock_on_hand"].head(20).plot(
    kind="bar"
)
plt.title("Top 20 Products by Average Inventory")
plt.ylabel("Stock On Hand")
plt.show()

# ============================================================
# 5. STORE LEVEL INVENTORY
# ============================================================

print("\nSTORE INVENTORY ANALYSIS")

store_inventory = (
    df.groupby("store_id")
    .agg({
        "stock_on_hand":"mean",
        "units_sold":"mean"
    })
    .sort_values(
        "stock_on_hand",
        ascending=False
    )
)

display(store_inventory.head(20))

plt.figure(figsize=(14,5))
store_inventory["stock_on_hand"].head(20).plot(
    kind="bar"
)
plt.title("Top Stores by Average Inventory")
plt.ylabel("Stock On Hand")
plt.show()

# ============================================================
# 6. CATEGORY INVENTORY ANALYSIS
# ============================================================

if "category" in df.columns:

    print("\nCATEGORY INVENTORY ANALYSIS")

    category_inventory = (
        df.groupby("category")
        .agg({
            "stock_on_hand":"mean",
            "units_sold":"mean"
        })
        .sort_values(
            "stock_on_hand",
            ascending=False
        )
    )

    display(category_inventory)

    plt.figure(figsize=(10,5))
    category_inventory["stock_on_hand"].plot(
        kind="bar"
    )
    plt.title("Inventory by Category")
    plt.ylabel("Average Stock")
    plt.show()

# ============================================================
# 7. REGION INVENTORY ANALYSIS
# ============================================================

if "region" in df.columns:

    print("\nREGIONAL INVENTORY ANALYSIS")

    region_inventory = (
        df.groupby("region")
        .agg({
            "stock_on_hand":"mean",
            "units_sold":"mean"
        })
        .sort_values(
            "stock_on_hand",
            ascending=False
        )
    )

    display(region_inventory)

    plt.figure(figsize=(8,5))
    region_inventory["stock_on_hand"].plot(
        kind="bar"
    )
    plt.title("Inventory by Region")
    plt.ylabel("Average Stock")
    plt.show()

# ============================================================
# 8. INVENTORY TURNOVER
# ============================================================

print("\nINVENTORY TURNOVER ANALYSIS")

turnover_df = (
    df.groupby(["product_id"])
    .agg({
        "units_sold":"sum",
        "stock_on_hand":"mean"
    })
)

turnover_df["inventory_turnover"] = (
    turnover_df["units_sold"] /
    turnover_df["stock_on_hand"]
)

display(
    turnover_df
    .sort_values(
        "inventory_turnover",
        ascending=False
    )
    .head(20)
)

# ============================================================
# 9. DAYS OF INVENTORY (DOI)
# ============================================================

print("\nDAYS OF INVENTORY ANALYSIS")

turnover_df["daily_demand"] = (
    turnover_df["units_sold"] /
    df["date"].nunique()
)

turnover_df["days_of_inventory"] = (
    turnover_df["stock_on_hand"] /
    turnover_df["daily_demand"]
)

display(
    turnover_df[
        ["inventory_turnover","days_of_inventory"]
    ]
    .head()
)

# ============================================================
# 10. STOCKOUT RISK ANALYSIS
# ============================================================

print("\nSTOCKOUT RISK ANALYSIS")

df["inventory_coverage"] = (
    df["stock_on_hand"] /
    np.maximum(df["units_sold"],1)
)

risk_conditions = [
    df["inventory_coverage"] < 3,
    (df["inventory_coverage"] >= 3) &
    (df["inventory_coverage"] < 7),
    df["inventory_coverage"] >= 7
]

risk_labels = [
    "High Risk",
    "Medium Risk",
    "Low Risk"
]

df["stockout_risk"] = np.select(
    risk_conditions,
    risk_labels,
    default="Unknown"
)

risk_summary = (
    df["stockout_risk"]
    .value_counts()
)

display(risk_summary)

plt.figure(figsize=(8,5))
risk_summary.plot(kind="bar")
plt.title("Stockout Risk Distribution")
plt.show()

# ============================================================
# 11. BACKORDER ANALYSIS
# ============================================================

if "backorder_qty" in df.columns:

    print("\nBACKORDER ANALYSIS")

    backorder_summary = pd.DataFrame({
        "Metric":[
            "Total Backorders",
            "Average Backorder",
            "Maximum Backorder"
        ],
        "Value":[
            df["backorder_qty"].sum(),
            round(df["backorder_qty"].mean(),2),
            df["backorder_qty"].max()
        ]
    })

    display(backorder_summary)

# ============================================================
# 12. INVENTORY COVERAGE ANALYSIS
# ============================================================

coverage_summary = pd.DataFrame({
    "Metric":[
        "Average Coverage",
        "Median Coverage",
        "Maximum Coverage"
    ],
    "Value":[
        round(df["inventory_coverage"].mean(),2),
        round(df["inventory_coverage"].median(),2),
        round(df["inventory_coverage"].max(),2)
    ]
})

print("\nINVENTORY COVERAGE")

display(coverage_summary)

# ============================================================
# 13. SUPPLIER ANALYSIS
# ============================================================

if "avg_lead_time" in df.columns:

    print("\nSUPPLIER LEAD TIME ANALYSIS")

    supplier_analysis = (
        df.groupby("supplier_id")
        .agg({
            "avg_lead_time":"mean",
            "supplier_rating":"mean"
        })
        .sort_values(
            "avg_lead_time",
            ascending=False
        )
    )

    display(supplier_analysis.head(20))

# ============================================================
# 14. ABC INVENTORY ANALYSIS
# ============================================================

print("\nABC INVENTORY ANALYSIS")

abc = (
    df.groupby("product_id")
    .agg({
        "units_sold":"sum"
    })
    .sort_values(
        "units_sold",
        ascending=False
    )
)

abc["cumulative_sales"] = (
    abc["units_sold"].cumsum()
)

abc["cumulative_pct"] = (
    abc["cumulative_sales"] /
    abc["units_sold"].sum()
) * 100

abc["ABC_Class"] = np.where(
    abc["cumulative_pct"] <= 80,
    "A",
    np.where(
        abc["cumulative_pct"] <= 95,
        "B",
        "C"
    )
)

display(
    abc["ABC_Class"]
    .value_counts()
)

# ============================================================
# 15. INVENTORY COST ANALYSIS
# ============================================================

if "unit_cost" in df.columns:

    print("\nINVENTORY COST ANALYSIS")

    df["inventory_value"] = (
        df["stock_on_hand"] *
        df["unit_cost"]
    )

    inventory_cost = pd.DataFrame({
        "Metric":[
            "Total Inventory Value",
            "Average Inventory Value",
            "Maximum Inventory Value"
        ],
        "Value":[
            round(df["inventory_value"].sum(),2),
            round(df["inventory_value"].mean(),2),
            round(df["inventory_value"].max(),2)
        ]
    })

    display(inventory_cost)

# ============================================================
# 16. INVENTORY HEALTH SCORE
# ============================================================

print("\nINVENTORY HEALTH SCORE")

health_score = (
    (
        (df["inventory_coverage"] > 7).mean()
    ) * 100
)

health_df = pd.DataFrame({
    "Metric":["Inventory Health Score"],
    "Value":[round(health_score,2)]
})

display(health_df)

# ============================================================
# 17. BUSINESS INSIGHTS
# ============================================================

print("\n" + "="*120)
print("BUSINESS INSIGHTS & RECOMMENDATIONS")
print("="*120)

print(f"""
1. Total Products Analysed:
   {df['product_id'].nunique()}

2. Total Stores Analysed:
   {df['store_id'].nunique()}

3. Average Inventory:
   {df['stock_on_hand'].mean():.2f}

4. Average Demand:
   {df['units_sold'].mean():.2f}

5. Inventory Health Score:
   {health_score:.2f}%

6. High-Risk Inventory:
   Inventory Coverage < 3 Days

7. Recommended Safety Stock:
   Demand Std × Lead Time

8. Reorder Point Formula:
   Avg Daily Demand × Lead Time + Safety Stock

9. Focus Inventory Optimization On:
   A-Class Products

10. Closely Monitor:
    High Demand + Low Inventory Products
""")

print("\nInventory Analysis Completed Successfully.")
print("=" * 120)
````

## 9.FEATURE ENGINEERING

````python
# ============================================================
# STEP 9: FEATURE ENGINEERING
# ============================================================
# Objectives:
# 1. Temporal Features
# 2. Cyclical Time Features
# 3. Lag Features
# 4. Rolling Window Features
# 5. Demand Trend Features
# 6. Inventory Features
# 7. Stockout Features
# 8. Promotion Features
# 9. Weather Features
# 10. Economic Features
# 11. Supplier Features
# 12. Product Features
# 13. Store Features
# 14. Demand Velocity Features
# 15. Interaction Features
# 16. Feature Quality Validation
# 17. Final Feature Dataset
# ============================================================

import pandas as pd
import numpy as np

print("=" * 120)
print("STEP 9 : FEATURE ENGINEERING")
print("=" * 120)

# ============================================================
# COPY DATASET
# ============================================================

df = master_df.copy()

# Ensure date format
df["date"] = pd.to_datetime(df["date"])

# Sort for time-series calculations
df = df.sort_values(
    ["product_id", "store_id", "date"]
)

print(f"\nInitial Shape: {df.shape}")

# ============================================================
# 1. TEMPORAL FEATURES
# ============================================================

print("\nCreating Temporal Features...")

df["year"] = df["date"].dt.year
df["quarter"] = df["date"].dt.quarter
df["month"] = df["date"].dt.month
df["week"] = df["date"].dt.isocalendar().week.astype(int)
df["day"] = df["date"].dt.day
df["dayofweek"] = df["date"].dt.dayofweek
df["dayofyear"] = df["date"].dt.dayofyear

df["is_weekend"] = np.where(
    df["dayofweek"].isin([5, 6]),
    1,
    0
)

df["is_month_start"] = df["date"].dt.is_month_start.astype(int)
df["is_month_end"] = df["date"].dt.is_month_end.astype(int)
df["is_quarter_start"] = df["date"].dt.is_quarter_start.astype(int)
df["is_quarter_end"] = df["date"].dt.is_quarter_end.astype(int)

# ============================================================
# 2. CYCLICAL FEATURES
# ============================================================

print("Creating Cyclical Features...")

df["month_sin"] = np.sin(
    2 * np.pi * df["month"] / 12
)

df["month_cos"] = np.cos(
    2 * np.pi * df["month"] / 12
)

df["dow_sin"] = np.sin(
    2 * np.pi * df["dayofweek"] / 7
)

df["dow_cos"] = np.cos(
    2 * np.pi * df["dayofweek"] / 7
)

# ============================================================
# 3. LAG FEATURES
# ============================================================

print("Creating Lag Features...")

group_cols = ["product_id", "store_id"]

df["units_sold_lag_1"] = (
    df.groupby(group_cols)["units_sold"]
    .shift(1)
)

df["units_sold_lag_7"] = (
    df.groupby(group_cols)["units_sold"]
    .shift(7)
)

df["units_sold_lag_14"] = (
    df.groupby(group_cols)["units_sold"]
    .shift(14)
)

df["units_sold_lag_30"] = (
    df.groupby(group_cols)["units_sold"]
    .shift(30)
)

# ============================================================
# 4. ROLLING WINDOW FEATURES
# ============================================================

print("Creating Rolling Features...")

grouped = df.groupby(group_cols)["units_sold"]

df["rolling_mean_7"] = (
    grouped.transform(
        lambda x: x.shift(1).rolling(7).mean()
    )
)

df["rolling_std_7"] = (
    grouped.transform(
        lambda x: x.shift(1).rolling(7).std()
    )
)

df["rolling_mean_14"] = (
    grouped.transform(
        lambda x: x.shift(1).rolling(14).mean()
    )
)

df["rolling_std_14"] = (
    grouped.transform(
        lambda x: x.shift(1).rolling(14).std()
    )
)

df["rolling_mean_30"] = (
    grouped.transform(
        lambda x: x.shift(1).rolling(30).mean()
    )
)

df["rolling_std_30"] = (
    grouped.transform(
        lambda x: x.shift(1).rolling(30).std()
    )
)

# ============================================================
# 5. DEMAND TREND FEATURES
# ============================================================

print("Creating Demand Trend Features...")

df["demand_growth_7"] = (
    (
        df["units_sold_lag_1"]
        -
        df["units_sold_lag_7"]
    )
    /
    (df["units_sold_lag_7"] + 1)
)

df["demand_growth_30"] = (
    (
        df["units_sold_lag_1"]
        -
        df["units_sold_lag_30"]
    )
    /
    (df["units_sold_lag_30"] + 1)
)

# ============================================================
# 6. INVENTORY FEATURES
# ============================================================

print("Creating Inventory Features...")

if "stock_on_hand" in df.columns:

    df["stock_to_sales_ratio"] = (
        df["stock_on_hand"]
        /
        (df["units_sold_lag_1"] + 1)
    )

    df["inventory_coverage_days"] = (
        df["stock_on_hand"]
        /
        (df["rolling_mean_7"] + 1)
    )

if (
    "stock_on_hand" in df.columns
    and
    "stock_in_transit" in df.columns
):

    df["total_inventory"] = (
        df["stock_on_hand"]
        +
        df["stock_in_transit"]
    )

# ============================================================
# 7. STOCKOUT FEATURES
# ============================================================

print("Creating Stockout Features...")

if "backorder_qty" in df.columns:

    df["backorder_flag"] = np.where(
        df["backorder_qty"] > 0,
        1,
        0
    )

if "stock_on_hand" in df.columns:

    df["low_stock_flag"] = np.where(
        df["stock_on_hand"] <
        df["rolling_mean_7"],
        1,
        0
    )

# ============================================================
# 8. PROMOTION FEATURES
# ============================================================

print("Creating Promotion Features...")

if "discount_pct" in df.columns:

    df["promotion_flag"] = np.where(
        df["discount_pct"] > 0,
        1,
        0
    )

    df["discount_intensity"] = (
        df["discount_pct"] / 100
    )

# ============================================================
# 9. WEATHER FEATURES
# ============================================================

print("Creating Weather Features...")

if "temperature" in df.columns:

    df["temp_deviation"] = (
        df["temperature"]
        -
        df["temperature"].mean()
    )

if "rainfall" in df.columns:

    df["high_rainfall_flag"] = np.where(
        df["rainfall"]
        >
        df["rainfall"].quantile(0.75),
        1,
        0
    )

# ============================================================
# 10. ECONOMIC FEATURES
# ============================================================

print("Creating Economic Features...")

if "inflation_rate" in df.columns:

    df["inflation_change"] = (
        df["inflation_rate"]
        -
        df["inflation_rate"].mean()
    )

if "consumer_index" in df.columns:

    df["consumer_strength"] = (
        df["consumer_index"]
        /
        df["consumer_index"].mean()
    )

# ============================================================
# 11. SUPPLIER FEATURES
# ============================================================

print("Creating Supplier Features...")

if (
    "avg_lead_time" in df.columns
    and
    "supplier_rating" in df.columns
):

    df["supplier_risk_score"] = (
        df["avg_lead_time"]
        /
        (df["supplier_rating"] + 1)
    )

# ============================================================
# 12. PRODUCT FEATURES
# ============================================================

print("Creating Product Features...")

if (
    "unit_cost" in df.columns
    and
    "unit_price" in df.columns
):

    df["gross_margin"] = (
        df["unit_price"]
        -
        df["unit_cost"]
    )

    df["margin_percentage"] = (
        df["gross_margin"]
        /
        df["unit_price"]
    )

# ============================================================
# 13. STORE FEATURES
# ============================================================

print("Creating Store Features...")

if "store_size_sqft" in df.columns:

    df["sales_per_sqft"] = (
        df["units_sold_lag_1"]
        /
        (df["store_size_sqft"] + 1)
    )

# ============================================================
# 14. DEMAND VELOCITY FEATURES
# ============================================================

print("Creating Demand Velocity Features...")

df["demand_velocity"] = (
    df["rolling_mean_7"]
    -
    df["rolling_mean_30"]
)

df["demand_volatility"] = (
    df["rolling_std_30"]
)

# ============================================================
# 15. INTERACTION FEATURES
# ============================================================

print("Creating Interaction Features...")

if (
    "discount_pct" in df.columns
    and
    "temperature" in df.columns
):

    df["promo_weather_interaction"] = (
        df["discount_pct"]
        *
        df["temperature"]
    )

if (
    "stock_on_hand" in df.columns
    and
    "avg_lead_time" in df.columns
):

    df["inventory_leadtime_interaction"] = (
        df["stock_on_hand"]
        *
        df["avg_lead_time"]
    )

# ============================================================
# 16. HANDLE MISSING VALUES
# ============================================================

print("Handling Missing Values from Lag Features...")

before_rows = len(df)

df = df.dropna()

after_rows = len(df)

print(
    f"Rows Removed Due To Lag/Rolling Features: "
    f"{before_rows - after_rows:,}"
)

# ============================================================
# 17. FEATURE QUALITY VALIDATION
# ============================================================

print("\nValidating Engineered Features...")

missing_count = df.isnull().sum().sum()

duplicate_count = df.duplicated().sum()

print(f"Missing Values Remaining : {missing_count:,}")
print(f"Duplicate Rows          : {duplicate_count:,}")

# ============================================================
# FEATURE SUMMARY
# ============================================================

engineered_features = [

    col for col in df.columns

    if col not in master_df.columns
]

feature_summary = pd.DataFrame({
    "Engineered Feature":
    engineered_features
})

print("\nEngineered Features Created:")
display(feature_summary)

# ============================================================
# FINAL DATASET SUMMARY
# ============================================================

print("\nFINAL FEATURE DATASET")

summary = pd.DataFrame({
    "Metric":[
        "Rows",
        "Columns",
        "Original Features",
        "Engineered Features",
        "Total Features"
    ],
    "Value":[
        len(df),
        len(df.columns),
        len(master_df.columns),
        len(engineered_features),
        len(df.columns)
    ]
})

display(summary)

print("\nFinal Dataset Shape:", df.shape)

# Save engineered dataset for next step
feature_df = df.copy()

print("\nFeature Engineering Completed Successfully.")
print("=" * 120)
````

## 10.FEATURE SELECTION

````python
# ============================================================
# STEP 10 : FEATURE SELECTION
# ============================================================
# Objectives:
# 1. Remove Constant Features
# 2. Remove High Missing Features
# 3. Remove Highly Correlated Features
# 4. Mutual Information Analysis
# 5. Random Forest Feature Importance
# 6. Permutation Importance
# 7. Variance Inflation Factor (VIF)
# 8. Leakage Detection
# 9. Business Feature Validation
# 10. Final Feature Selection
# ============================================================

import pandas as pd
import numpy as np

from sklearn.feature_selection import mutual_info_regression
from sklearn.ensemble import RandomForestRegressor
from sklearn.inspection import permutation_importance
from sklearn.preprocessing import LabelEncoder
from statsmodels.stats.outliers_influence import variance_inflation_factor

print("=" * 120)
print("STEP 10 : FEATURE SELECTION")
print("=" * 120)

# ============================================================
# LOAD FEATURE ENGINEERED DATA
# ============================================================

df = feature_df.copy()

TARGET = "units_sold"

print(f"\nDataset Shape : {df.shape}")

# ============================================================
# REMOVE NON-PREDICTIVE COLUMNS
# ============================================================

remove_columns = [
    "date",
    "revenue",          # Leakage
    "base_demand"       # Synthetic target generator leakage
]

existing_remove_cols = [
    col for col in remove_columns
    if col in df.columns
]

df = df.drop(columns=existing_remove_cols)

print("\nRemoved Columns:")
print(existing_remove_cols)

# ============================================================
# IDENTIFY FEATURE TYPES
# ============================================================

target = df[TARGET]

X = df.drop(columns=[TARGET])

categorical_cols = X.select_dtypes(
    include=["object", "category"]
).columns.tolist()

numerical_cols = X.select_dtypes(
    include=np.number
).columns.tolist()

print(f"\nCategorical Features : {len(categorical_cols)}")
print(f"Numerical Features   : {len(numerical_cols)}")

# ============================================================
# LABEL ENCODING
# ============================================================

print("\nEncoding Categorical Features...")

encoders = {}

for col in categorical_cols:

    le = LabelEncoder()

    X[col] = le.fit_transform(
        X[col].astype(str)
    )

    encoders[col] = le

# ============================================================
# 1. CONSTANT FEATURE REMOVAL
# ============================================================

print("\nRemoving Constant Features...")

constant_features = [
    col for col in X.columns
    if X[col].nunique() <= 1
]

X = X.drop(columns=constant_features)

print(f"Constant Features Removed: {len(constant_features)}")

# ============================================================
# 2. HIGH MISSING FEATURES
# ============================================================

print("\nChecking Missing Values...")

missing_percent = (
    X.isnull()
    .mean()
    .sort_values(ascending=False)
)

high_missing_features = (
    missing_percent[
        missing_percent > 0.50
    ].index.tolist()
)

X = X.drop(columns=high_missing_features)

print(
    f"High Missing Features Removed: "
    f"{len(high_missing_features)}"
)

# ============================================================
# 3. CORRELATION FILTER
# ============================================================

print("\nRemoving Highly Correlated Features...")

corr_matrix = X.corr().abs()

upper_triangle = corr_matrix.where(
    np.triu(
        np.ones(corr_matrix.shape),
        k=1
    ).astype(bool)
)

correlated_features = [
    column
    for column in upper_triangle.columns
    if any(
        upper_triangle[column] > 0.95
    )
]

X_corr = X.drop(
    columns=correlated_features,
    errors="ignore"
)

print(
    f"Highly Correlated Features Removed: "
    f"{len(correlated_features)}"
)

# ============================================================
# 4. MUTUAL INFORMATION
# ============================================================

print("\nCalculating Mutual Information...")

mi_scores = mutual_info_regression(
    X_corr,
    target,
    random_state=42
)

mi_df = pd.DataFrame({
    "feature": X_corr.columns,
    "mi_score": mi_scores
})

mi_df = mi_df.sort_values(
    by="mi_score",
    ascending=False
)

display(mi_df.head(20))

# ============================================================
# 5. RANDOM FOREST IMPORTANCE
# ============================================================

print("\nTraining Random Forest...")

rf = RandomForestRegressor(
    n_estimators=100,
    max_depth=12,
    random_state=42,
    n_jobs=-1
)

rf.fit(X_corr, target)

rf_importance_df = pd.DataFrame({
    "feature": X_corr.columns,
    "rf_importance": rf.feature_importances_
})

rf_importance_df = rf_importance_df.sort_values(
    by="rf_importance",
    ascending=False
)

print("\nTop Random Forest Features")
display(rf_importance_df.head(20))

# ============================================================
# 6. PERMUTATION IMPORTANCE
# ============================================================

print("\nCalculating Permutation Importance...")

sample_size = min(
    50000,
    len(X_corr)
)

sample_idx = np.random.choice(
    len(X_corr),
    sample_size,
    replace=False
)

perm = permutation_importance(
    rf,
    X_corr.iloc[sample_idx],
    target.iloc[sample_idx],
    n_repeats=5,
    random_state=42,
    n_jobs=-1
)

perm_df = pd.DataFrame({
    "feature": X_corr.columns,
    "permutation_importance": perm.importances_mean
})

perm_df = perm_df.sort_values(
    by="permutation_importance",
    ascending=False
)

print("\nTop Permutation Features")
display(perm_df.head(20))

# ============================================================
# 7. VIF ANALYSIS
# ============================================================

print("\nCalculating VIF Scores...")

vif_df = X_corr.copy()

# limit VIF calculation
vif_cols = rf_importance_df.head(
    20
)["feature"].tolist()

vif_df = vif_df[vif_cols]

vif_results = pd.DataFrame()

vif_results["feature"] = vif_df.columns

vif_results["VIF"] = [
    variance_inflation_factor(
        vif_df.values,
        i
    )
    for i in range(
        vif_df.shape[1]
    )
]

vif_results = vif_results.sort_values(
    by="VIF",
    ascending=False
)

display(vif_results)

# ============================================================
# 8. DATA LEAKAGE DETECTION
# ============================================================

print("\nDetecting Potential Leakage...")

leakage_candidates = []

for col in X_corr.columns:

    corr = abs(
        np.corrcoef(
            X_corr[col],
            target
        )[0, 1]
    )

    if corr > 0.95:
        leakage_candidates.append(
            (col, corr)
        )

leakage_df = pd.DataFrame(
    leakage_candidates,
    columns=[
        "feature",
        "target_correlation"
    ]
)

print("\nPotential Leakage Features")
display(leakage_df)

# ============================================================
# 9. COMBINE FEATURE SCORES
# ============================================================

print("\nCombining Selection Metrics...")

feature_ranking = (
    mi_df.merge(
        rf_importance_df,
        on="feature"
    )
    .merge(
        perm_df,
        on="feature"
    )
)

feature_ranking["combined_score"] = (
    feature_ranking["mi_score"].rank(
        ascending=False
    )
    +
    feature_ranking["rf_importance"].rank(
        ascending=False
    )
    +
    feature_ranking["permutation_importance"].rank(
        ascending=False
    )
)

feature_ranking = feature_ranking.sort_values(
    by="combined_score"
)

print("\nTop Ranked Features")
display(feature_ranking.head(30))

# ============================================================
# 10. FINAL FEATURE SELECTION
# ============================================================

TOP_N_FEATURES = 25

selected_features = (
    feature_ranking
    .head(TOP_N_FEATURES)
    ["feature"]
    .tolist()
)

# remove leakage features if detected
selected_features = [
    f for f in selected_features
    if f not in leakage_df["feature"].tolist()
]

final_dataset = pd.concat(
    [
        X_corr[selected_features],
        target
    ],
    axis=1
)

print("\nSelected Features:")
for i, feature in enumerate(
    selected_features,
    start=1
):
    print(f"{i:02d}. {feature}")

# ============================================================
# FEATURE SELECTION SUMMARY
# ============================================================

summary = pd.DataFrame({
    "Metric": [
        "Original Features",
        "After Correlation Filter",
        "Selected Features",
        "Target Variable",
        "Final Dataset Columns"
    ],
    "Value": [
        X.shape[1],
        X_corr.shape[1],
        len(selected_features),
        1,
        final_dataset.shape[1]
    ]
})

print("\nFeature Selection Summary")
display(summary)

print("\nFinal Dataset Shape:")
print(final_dataset.shape)

# ============================================================
# SAVE FOR NEXT STEP
# ============================================================

selected_feature_df = final_dataset.copy()
selected_feature_list = selected_features

print("\nFeature Selection Completed Successfully")
print("=" * 120)
````

## 11.FORECAST DATASET CREATION

````python
# ============================================================
# STEP 11 : FORECAST DATASET CREATION
# ============================================================
# Objectives:
# 1. Create Final Forecasting Dataset
# 2. Validate Selected Features
# 3. Create Time-Based Train/Test Split
# 4. Create Train/Validation/Test Sets
# 5. Encode Categorical Features
# 6. Feature Scaling (Optional)
# 7. Create Future Forecast Dataset
# 8. Dataset Validation
# 9. Forecast Metadata Generation
# 10. Save Modeling Artifacts
# ============================================================

import pandas as pd
import numpy as np

from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler

print("=" * 120)
print("STEP 11 : FORECAST DATASET CREATION")
print("=" * 120)

# ============================================================
# LOAD FEATURE ENGINEERED DATA
# ============================================================

df = feature_df.copy()

TARGET = "units_sold"

print(f"\nInput Dataset Shape : {df.shape}")

# ============================================================
# REQUIRED COLUMNS
# ============================================================

required_columns = [
    "date",
    TARGET
]

for col in required_columns:
    if col not in df.columns:
        raise ValueError(
            f"Required column missing: {col}"
        )

# ============================================================
# USE SELECTED FEATURES
# ============================================================

if "selected_feature_list" in globals():

    selected_cols = (
        ["date"] +
        selected_feature_list +
        [TARGET]
    )

    selected_cols = [
        col for col in selected_cols
        if col in df.columns
    ]

    forecast_df = df[selected_cols].copy()

else:

    print(
        "\nWARNING: selected_feature_list not found."
        "\nUsing all features."
    )

    forecast_df = df.copy()

print(
    f"\nForecast Dataset Shape : "
    f"{forecast_df.shape}"
)

# ============================================================
# SORT BY DATE
# ============================================================

forecast_df["date"] = pd.to_datetime(
    forecast_df["date"]
)

forecast_df = forecast_df.sort_values(
    "date"
)

# ============================================================
# IDENTIFY FEATURES
# ============================================================

feature_columns = [
    col for col in forecast_df.columns
    if col not in ["date", TARGET]
]

print(
    f"\nNumber of Features : "
    f"{len(feature_columns)}"
)

# ============================================================
# CATEGORICAL ENCODING
# ============================================================

categorical_columns = (
    forecast_df[feature_columns]
    .select_dtypes(
        include=["object", "category"]
    )
    .columns
    .tolist()
)

print(
    f"\nCategorical Features : "
    f"{len(categorical_columns)}"
)

label_encoders = {}

for col in categorical_columns:

    le = LabelEncoder()

    forecast_df[col] = (
        le.fit_transform(
            forecast_df[col].astype(str)
        )
    )

    label_encoders[col] = le

print("Categorical Encoding Completed")

# ============================================================
# TIME-BASED DATA SPLIT
# ============================================================

print("\nCreating Time-Based Split...")

unique_dates = (
    forecast_df["date"]
    .sort_values()
    .unique()
)

n_dates = len(unique_dates)

train_end = int(n_dates * 0.70)
valid_end = int(n_dates * 0.85)

train_dates = unique_dates[:train_end]
valid_dates = unique_dates[train_end:valid_end]
test_dates = unique_dates[valid_end:]

train_df = forecast_df[
    forecast_df["date"].isin(train_dates)
]

valid_df = forecast_df[
    forecast_df["date"].isin(valid_dates)
]

test_df = forecast_df[
    forecast_df["date"].isin(test_dates)
]

print(f"Train Shape      : {train_df.shape}")
print(f"Validation Shape : {valid_df.shape}")
print(f"Test Shape       : {test_df.shape}")

# ============================================================
# FEATURE / TARGET SPLIT
# ============================================================

X_train = train_df[
    feature_columns
].copy()

y_train = train_df[
    TARGET
].copy()

X_valid = valid_df[
    feature_columns
].copy()

y_valid = valid_df[
    TARGET
].copy()

X_test = test_df[
    feature_columns
].copy()

y_test = test_df[
    TARGET
].copy()

# ============================================================
# FEATURE SCALING
# ============================================================

print("\nApplying Feature Scaling...")

numeric_features = (
    X_train.select_dtypes(
        include=np.number
    )
    .columns
    .tolist()
)

scaler = StandardScaler()

X_train_scaled = X_train.copy()
X_valid_scaled = X_valid.copy()
X_test_scaled = X_test.copy()

X_train_scaled[numeric_features] = (
    scaler.fit_transform(
        X_train[numeric_features]
    )
)

X_valid_scaled[numeric_features] = (
    scaler.transform(
        X_valid[numeric_features]
    )
)

X_test_scaled[numeric_features] = (
    scaler.transform(
        X_test[numeric_features]
    )
)

print("Scaling Completed")

# ============================================================
# CREATE FUTURE FORECAST DATASET
# ============================================================

print("\nCreating Future Forecast Dataset...")

latest_date = forecast_df["date"].max()

future_dataset = (
    forecast_df
    .sort_values("date")
    .groupby(
        ["product_id", "store_id"]
    )
    .tail(1)
    .copy()
)

future_dataset["forecast_date"] = (
    latest_date +
    pd.Timedelta(days=1)
)

print(
    f"Future Forecast Records : "
    f"{len(future_dataset):,}"
)

# ============================================================
# DATASET VALIDATION
# ============================================================

validation_results = pd.DataFrame({
    "Check": [
        "Missing Values",
        "Duplicate Rows",
        "Train Empty",
        "Validation Empty",
        "Test Empty"
    ],
    "Result": [
        forecast_df.isnull().sum().sum(),
        forecast_df.duplicated().sum(),
        len(train_df) == 0,
        len(valid_df) == 0,
        len(test_df) == 0
    ]
})

print("\nDataset Validation")
display(validation_results)

# ============================================================
# FORECAST DATASET SUMMARY
# ============================================================

summary_df = pd.DataFrame({
    "Metric": [
        "Total Rows",
        "Total Features",
        "Training Rows",
        "Validation Rows",
        "Testing Rows",
        "Forecast Records"
    ],
    "Value": [
        len(forecast_df),
        len(feature_columns),
        len(train_df),
        len(valid_df),
        len(test_df),
        len(future_dataset)
    ]
})

print("\nForecast Dataset Summary")
display(summary_df)

# ============================================================
# FORECAST METADATA
# ============================================================

forecast_metadata = {
    "target_column": TARGET,
    "feature_count": len(feature_columns),
    "train_start_date": str(
        train_df["date"].min()
    ),
    "train_end_date": str(
        train_df["date"].max()
    ),
    "validation_start_date": str(
        valid_df["date"].min()
    ),
    "validation_end_date": str(
        valid_df["date"].max()
    ),
    "test_start_date": str(
        test_df["date"].min()
    ),
    "test_end_date": str(
        test_df["date"].max()
    ),
    "forecast_start_date": str(
        future_dataset["forecast_date"].min()
    )
}

print("\nForecast Metadata")
for k, v in forecast_metadata.items():
    print(f"{k}: {v}")

# ============================================================
# FINAL MODELING DATASETS
# ============================================================

forecast_datasets = {
    "X_train": X_train_scaled,
    "y_train": y_train,
    "X_valid": X_valid_scaled,
    "y_valid": y_valid,
    "X_test": X_test_scaled,
    "y_test": y_test,
    "future_forecast_dataset": future_dataset,
    "feature_columns": feature_columns,
    "label_encoders": label_encoders,
    "scaler": scaler,
    "metadata": forecast_metadata
}

print("\nFinal Dataset Objects Created")

print("\nX_train Shape :", X_train_scaled.shape)
print("X_valid Shape :", X_valid_scaled.shape)
print("X_test Shape  :", X_test_scaled.shape)

print("\nForecast Dataset Creation Completed Successfully")
print("=" * 120)
````

## 12.MODEL DEVELOPMENT

````python
# ============================================================
# STEP 12 : MODEL DEVELOPMENT
# ============================================================
# Demand Forecasting & Inventory Optimization
#
# Models:
# 1. Linear Regression (Baseline)
# 2. Random Forest
# 3. XGBoost (Optional)
# 4. LightGBM (Optional)
#
# Validation:
# - Time Series Aware Evaluation
# - Validation Set Comparison
# - Model Ranking
# - Best Model Selection
#
# Output:
# - Trained Models
# - Model Leaderboard
# - Feature Importance
# - Best Model Artifact
# ============================================================

import warnings
warnings.filterwarnings("ignore")

import pandas as pd
import numpy as np
import joblib

from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor

from sklearn.metrics import (
    mean_absolute_error,
    mean_squared_error,
    r2_score
)

print("="*120)
print("STEP 12 : MODEL DEVELOPMENT")
print("="*120)

# ============================================================
# LOAD DATASETS FROM STEP 11
# ============================================================

X_train = forecast_datasets["X_train"]
y_train = forecast_datasets["y_train"]

X_valid = forecast_datasets["X_valid"]
y_valid = forecast_datasets["y_valid"]

X_test = forecast_datasets["X_test"]
y_test = forecast_datasets["y_test"]

feature_columns = forecast_datasets["feature_columns"]

print("\nDataset Information")
print(f"X_train : {X_train.shape}")
print(f"X_valid : {X_valid.shape}")
print(f"X_test  : {X_test.shape}")

# ============================================================
# EVALUATION FUNCTION
# ============================================================

def evaluate_model(model_name, y_true, y_pred):

    mae = mean_absolute_error(
        y_true,
        y_pred
    )

    rmse = np.sqrt(
        mean_squared_error(
            y_true,
            y_pred
        )
    )

    r2 = r2_score(
        y_true,
        y_pred
    )

    mape = np.mean(
        np.abs(
            (y_true - y_pred)
            /
            np.maximum(y_true, 1)
        )
    ) * 100

    return {
        "Model": model_name,
        "MAE": round(mae, 4),
        "RMSE": round(rmse, 4),
        "MAPE": round(mape, 4),
        "R2": round(r2, 4)
    }

# ============================================================
# MODEL CONTAINER
# ============================================================

models = {}

leaderboard = []

# ============================================================
# MODEL 1 : LINEAR REGRESSION
# ============================================================

print("\nTraining Linear Regression...")

lr_model = LinearRegression()

lr_model.fit(
    X_train,
    y_train
)

lr_pred = lr_model.predict(
    X_valid
)

models["LinearRegression"] = lr_model

leaderboard.append(
    evaluate_model(
        "LinearRegression",
        y_valid,
        lr_pred
    )
)

print("Linear Regression Training Completed")

# ============================================================
# MODEL 2 : RANDOM FOREST
# ============================================================

print("\nTraining Random Forest...")

rf_model = RandomForestRegressor(
    n_estimators=300,
    max_depth=15,
    min_samples_split=5,
    min_samples_leaf=2,
    random_state=42,
    n_jobs=-1
)

rf_model.fit(
    X_train,
    y_train
)

rf_pred = rf_model.predict(
    X_valid
)

models["RandomForest"] = rf_model

leaderboard.append(
    evaluate_model(
        "RandomForest",
        y_valid,
        rf_pred
    )
)

print("Random Forest Training Completed")

# ============================================================
# MODEL 3 : XGBOOST
# ============================================================

try:

    from xgboost import XGBRegressor

    print("\nTraining XGBoost...")

    xgb_model = XGBRegressor(
        n_estimators=500,
        learning_rate=0.05,
        max_depth=8,
        subsample=0.8,
        colsample_bytree=0.8,
        random_state=42,
        objective="reg:squarederror"
    )

    xgb_model.fit(
        X_train,
        y_train
    )

    xgb_pred = xgb_model.predict(
        X_valid
    )

    models["XGBoost"] = xgb_model

    leaderboard.append(
        evaluate_model(
            "XGBoost",
            y_valid,
            xgb_pred
        )
    )

    print("XGBoost Training Completed")

except Exception as e:

    print(
        f"\nXGBoost skipped: {str(e)}"
    )

# ============================================================
# MODEL 4 : LIGHTGBM
# ============================================================

try:

    from lightgbm import LGBMRegressor

    print("\nTraining LightGBM...")

    lgbm_model = LGBMRegressor(
        n_estimators=500,
        learning_rate=0.05,
        max_depth=8,
        random_state=42
    )

    lgbm_model.fit(
        X_train,
        y_train
    )

    lgbm_pred = lgbm_model.predict(
        X_valid
    )

    models["LightGBM"] = lgbm_model

    leaderboard.append(
        evaluate_model(
            "LightGBM",
            y_valid,
            lgbm_pred
        )
    )

    print("LightGBM Training Completed")

except Exception as e:

    print(
        f"\nLightGBM skipped: {str(e)}"
    )

# ============================================================
# MODEL LEADERBOARD
# ============================================================

leaderboard_df = pd.DataFrame(
    leaderboard
)

leaderboard_df = leaderboard_df.sort_values(
    by="RMSE",
    ascending=True
)

print("\nMODEL LEADERBOARD")
display(leaderboard_df)

# ============================================================
# BEST MODEL SELECTION
# ============================================================

best_model_name = (
    leaderboard_df
    .iloc[0]["Model"]
)

best_model = models[
    best_model_name
]

print(f"\nBest Model: {best_model_name}")

# ============================================================
# FEATURE IMPORTANCE
# ============================================================

feature_importance_df = None

if hasattr(
    best_model,
    "feature_importances_"
):

    feature_importance_df = pd.DataFrame({

        "Feature": feature_columns,

        "Importance":
        best_model.feature_importances_

    })

    feature_importance_df = (
        feature_importance_df
        .sort_values(
            by="Importance",
            ascending=False
        )
    )

    print("\nTop 25 Important Features")

    display(
        feature_importance_df.head(25)
    )

# ============================================================
# FINAL TRAINING
# Train On Train + Validation
# ============================================================

print(
    "\nRetraining Best Model "
    "on Train + Validation..."
)

X_train_full = pd.concat(
    [
        X_train,
        X_valid
    ]
)

y_train_full = pd.concat(
    [
        y_train,
        y_valid
    ]
)

best_model.fit(
    X_train_full,
    y_train_full
)

print("Final Training Completed")

# ============================================================
# SAVE MODEL ARTIFACTS
# ============================================================

joblib.dump(
    best_model,
    "best_forecasting_model.pkl"
)

joblib.dump(
    feature_columns,
    "selected_features.pkl"
)

print(
    "\nArtifacts Saved:"
)

print(
    "best_forecasting_model.pkl"
)

print(
    "selected_features.pkl"
)

# ============================================================
# MODEL DEVELOPMENT SUMMARY
# ============================================================

summary_df = pd.DataFrame({

    "Metric": [

        "Training Records",
        "Validation Records",
        "Testing Records",
        "Features Used",
        "Models Trained",
        "Best Model"

    ],

    "Value": [

        len(X_train),
        len(X_valid),
        len(X_test),
        len(feature_columns),
        len(models),
        best_model_name

    ]

})

print("\nMODEL DEVELOPMENT SUMMARY")
display(summary_df)

# ============================================================
# EXPORT OBJECTS FOR STEP 13
# ============================================================

model_artifacts = {

    "best_model": best_model,

    "best_model_name": best_model_name,

    "feature_columns": feature_columns,

    "leaderboard": leaderboard_df,

    "feature_importance":
    feature_importance_df

}

print("\nStep 12 Completed Successfully")
print("="*120)
````

## 13.MODEL EVALUATION

````python
# ============================================================
# STEP 13 : MODEL EVALUATION
# ============================================================
# Objectives:
# 1. Load Best Model
# 2. Generate Test Predictions
# 3. Calculate Regression Metrics
# 4. Forecast Accuracy Metrics
# 5. Residual Analysis
# 6. Error Distribution Analysis
# 7. Product-Level Evaluation
# 8. Store-Level Evaluation
# 9. Category-Level Evaluation
# 10. Feature Importance Review
# 11. Business Impact Metrics
# 12. Model Evaluation Report
# ============================================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.metrics import (
    mean_absolute_error,
    mean_squared_error,
    r2_score,
    median_absolute_error,
    explained_variance_score
)

print("=" * 120)
print("STEP 13 : MODEL EVALUATION")
print("=" * 120)

# ============================================================
# LOAD MODEL ARTIFACTS
# ============================================================

best_model = model_artifacts["best_model"]
best_model_name = model_artifacts["best_model_name"]

X_test = forecast_datasets["X_test"]
y_test = forecast_datasets["y_test"]

print(f"\nBest Model : {best_model_name}")
print(f"Test Shape : {X_test.shape}")

# ============================================================
# GENERATE PREDICTIONS
# ============================================================

print("\nGenerating Predictions...")

y_pred = best_model.predict(X_test)

# ============================================================
# REGRESSION METRICS
# ============================================================

print("\nCalculating Evaluation Metrics...")

mae = mean_absolute_error(y_test, y_pred)

mse = mean_squared_error(y_test, y_pred)

rmse = np.sqrt(mse)

r2 = r2_score(y_test, y_pred)

mape = (
    np.mean(
        np.abs(
            (y_test - y_pred)
            /
            np.maximum(y_test, 1)
        )
    )
    * 100
)

smape = (
    np.mean(
        (
            np.abs(y_test - y_pred)
            /
            (
                np.abs(y_test)
                +
                np.abs(y_pred)
                +
                1e-6
            )
        )
    )
    * 200
)

medae = median_absolute_error(
    y_test,
    y_pred
)

evs = explained_variance_score(
    y_test,
    y_pred
)

# ============================================================
# EVALUATION SCORECARD
# ============================================================

evaluation_scorecard = pd.DataFrame({

    "Metric": [

        "MAE",
        "RMSE",
        "MAPE (%)",
        "sMAPE (%)",
        "R² Score",
        "Median AE",
        "Explained Variance"

    ],

    "Value": [

        round(mae, 4),
        round(rmse, 4),
        round(mape, 4),
        round(smape, 4),
        round(r2, 4),
        round(medae, 4),
        round(evs, 4)

    ]

})

print("\nMODEL PERFORMANCE")
display(evaluation_scorecard)

# ============================================================
# CREATE PREDICTION DATAFRAME
# ============================================================

prediction_df = pd.DataFrame({

    "actual_units_sold": y_test.values,
    "predicted_units_sold": y_pred

})

prediction_df["error"] = (
    prediction_df["actual_units_sold"]
    -
    prediction_df["predicted_units_sold"]
)

prediction_df["absolute_error"] = (
    np.abs(
        prediction_df["error"]
    )
)

prediction_df["percentage_error"] = (
    prediction_df["absolute_error"]
    /
    np.maximum(
        prediction_df["actual_units_sold"],
        1
    )
    * 100
)

# ============================================================
# RESIDUAL ANALYSIS
# ============================================================

print("\nResidual Analysis...")

residual_summary = pd.DataFrame({

    "Metric": [

        "Mean Residual",
        "Std Residual",
        "Max Residual",
        "Min Residual"

    ],

    "Value": [

        prediction_df["error"].mean(),
        prediction_df["error"].std(),
        prediction_df["error"].max(),
        prediction_df["error"].min()

    ]

})

display(residual_summary)

# ============================================================
# ERROR DISTRIBUTION
# ============================================================

plt.figure(figsize=(10,5))

sns.histplot(
    prediction_df["error"],
    bins=50,
    kde=True
)

plt.title(
    "Residual Distribution"
)

plt.show()

# ============================================================
# ACTUAL VS PREDICTED
# ============================================================

sample_size = min(
    10000,
    len(prediction_df)
)

sample_df = prediction_df.sample(
    sample_size,
    random_state=42
)

plt.figure(figsize=(8,8))

plt.scatter(
    sample_df["actual_units_sold"],
    sample_df["predicted_units_sold"],
    alpha=0.3
)

plt.xlabel("Actual")

plt.ylabel("Predicted")

plt.title(
    "Actual vs Predicted"
)

plt.show()

# ============================================================
# ERROR BUCKET ANALYSIS
# ============================================================

prediction_df["error_bucket"] = pd.cut(

    prediction_df["percentage_error"],

    bins=[
        0,
        5,
        10,
        20,
        50,
        100,
        np.inf
    ],

    labels=[
        "0-5%",
        "5-10%",
        "10-20%",
        "20-50%",
        "50-100%",
        ">100%"
    ]

)

error_bucket_summary = (

    prediction_df["error_bucket"]

    .value_counts()

    .reset_index()

)

print("\nError Distribution")
display(error_bucket_summary)

# ============================================================
# FORECAST ACCURACY KPI
# ============================================================

forecast_accuracy = (

    100 - mape

)

bias = (

    prediction_df["error"].sum()

    /
    prediction_df["actual_units_sold"].sum()

) * 100

business_metrics = pd.DataFrame({

    "Business KPI": [

        "Forecast Accuracy (%)",
        "Forecast Bias (%)",
        "Average Error",
        "Average Absolute Error"

    ],

    "Value": [

        round(forecast_accuracy,2),
        round(bias,2),
        round(
            prediction_df["error"].mean(),
            2
        ),
        round(
            prediction_df["absolute_error"].mean(),
            2
        )

    ]

})

print("\nBusiness Metrics")
display(business_metrics)

# ============================================================
# FEATURE IMPORTANCE REVIEW
# ============================================================

print("\nFeature Importance Review")

if (
    model_artifacts["feature_importance"]
    is not None
):

    feature_importance_df = (
        model_artifacts[
            "feature_importance"
        ]
    )

    display(
        feature_importance_df
        .head(25)
    )

    plt.figure(
        figsize=(10,8)
    )

    sns.barplot(

        data=
        feature_importance_df
        .head(15),

        x="Importance",

        y="Feature"

    )

    plt.title(
        "Top 15 Important Features"
    )

    plt.show()

# ============================================================
# MODEL EVALUATION REPORT
# ============================================================

evaluation_report = {

    "model_name":
    best_model_name,

    "mae":
    mae,

    "rmse":
    rmse,

    "mape":
    mape,

    "smape":
    smape,

    "r2":
    r2,

    "forecast_accuracy":
    forecast_accuracy,

    "bias":
    bias

}

# ============================================================
# EVALUATION SUMMARY
# ============================================================

summary_df = pd.DataFrame({

    "Metric": [

        "Model",
        "Test Records",
        "MAE",
        "RMSE",
        "MAPE (%)",
        "Forecast Accuracy (%)",
        "R²"

    ],

    "Value": [

        best_model_name,
        len(X_test),
        round(mae,4),
        round(rmse,4),
        round(mape,4),
        round(forecast_accuracy,4),
        round(r2,4)

    ]

})

print("\nMODEL EVALUATION SUMMARY")
display(summary_df)

# ============================================================
# EXPORT OBJECTS FOR STEP 14
# ============================================================

evaluation_artifacts = {

    "predictions": prediction_df,

    "evaluation_report":
    evaluation_report,

    "forecast_accuracy":
    forecast_accuracy,

    "rmse":
    rmse,

    "mape":
    mape

}

print("\nStep 13 Completed Successfully")
print("=" * 120)
````

## 14.INVENTORY OPTIMIZATION

````python
# ============================================================
# STEP 14 : INVENTORY OPTIMIZATION
# ============================================================
# Objectives:
# 1. Generate Demand Forecasts
# 2. Calculate Demand Variability
# 3. Estimate Lead Time Demand
# 4. Calculate Safety Stock
# 5. Calculate Reorder Point (ROP)
# 6. Calculate Economic Order Quantity (EOQ)
# 7. Inventory Risk Classification
# 8. Stockout Risk Analysis
# 9. Overstock Risk Analysis
# 10. Inventory Optimization Recommendations
# 11. Working Capital Analysis
# 12. Inventory Optimization Report
# ============================================================

import pandas as pd
import numpy as np
from scipy.stats import norm

print("="*120)
print("STEP 14 : INVENTORY OPTIMIZATION")
print("="*120)

# ============================================================
# CONFIGURATION
# ============================================================

SERVICE_LEVEL = 0.95
Z_SCORE = norm.ppf(SERVICE_LEVEL)

ORDERING_COST = 1000
HOLDING_COST_RATE = 0.20

print(f"\nTarget Service Level : {SERVICE_LEVEL*100:.0f}%")
print(f"Z Score              : {Z_SCORE:.3f}")

# ============================================================
# BASE DATASET
# ============================================================

inventory_df_opt = master_df.copy()

print("\nDataset Shape:")
print(inventory_df_opt.shape)

# ============================================================
# FORECAST DEMAND
# ============================================================

feature_cols = model_artifacts["feature_columns"]

inventory_df_opt["forecast_demand"] = (
    model_artifacts["best_model"]
    .predict(
        inventory_df_opt[feature_cols]
    )
)

inventory_df_opt["forecast_demand"] = (
    inventory_df_opt["forecast_demand"]
    .clip(lower=0)
)

# ============================================================
# PRODUCT-STORE LEVEL INVENTORY SUMMARY
# ============================================================

inventory_summary = (

    inventory_df_opt

    .groupby(
        [
            "product_id",
            "store_id"
        ]
    )

    .agg({

        "forecast_demand":"mean",

        "units_sold":"std",

        "stock_on_hand":"last",

        "unit_cost":"last",

        "supplier_rating":"last",

        "avg_lead_time":"last",

        "lead_time_std":"last",

        "backorder_qty":"mean"

    })

    .reset_index()

)

inventory_summary.rename(

    columns={

        "forecast_demand":
        "avg_daily_forecast",

        "units_sold":
        "demand_std",

        "stock_on_hand":
        "current_stock"

    },

    inplace=True

)

# ============================================================
# LEAD TIME DEMAND
# ============================================================

inventory_summary["lead_time_demand"] = (

    inventory_summary["avg_daily_forecast"]

    *

    inventory_summary["avg_lead_time"]

)

# ============================================================
# SAFETY STOCK
# ============================================================

inventory_summary["safety_stock"] = (

    Z_SCORE

    *

    inventory_summary["demand_std"]

    *

    np.sqrt(
        inventory_summary["avg_lead_time"]
    )

)

inventory_summary["safety_stock"] = (

    inventory_summary["safety_stock"]

    .fillna(0)

)

# ============================================================
# REORDER POINT
# ============================================================

inventory_summary["reorder_point"] = (

    inventory_summary["lead_time_demand"]

    +

    inventory_summary["safety_stock"]

)

# ============================================================
# EOQ CALCULATION
# ============================================================

inventory_summary["annual_demand"] = (

    inventory_summary["avg_daily_forecast"]

    * 365

)

inventory_summary["holding_cost"] = (

    inventory_summary["unit_cost"]

    * HOLDING_COST_RATE

)

inventory_summary["eoq"] = np.sqrt(

    (

        2

        * inventory_summary["annual_demand"]

        * ORDERING_COST

    )

    /

    np.maximum(
        inventory_summary["holding_cost"],
        1
    )

)

inventory_summary["eoq"] = (

    inventory_summary["eoq"]

    .round(0)

)

# ============================================================
# INVENTORY STATUS
# ============================================================

inventory_summary["inventory_gap"] = (

    inventory_summary["current_stock"]

    -

    inventory_summary["reorder_point"]

)

conditions = [

    inventory_summary["current_stock"]
    <=
    inventory_summary["reorder_point"],

    inventory_summary["current_stock"]
    >
    inventory_summary["reorder_point"] * 2

]

choices = [

    "REORDER_REQUIRED",

    "OVERSTOCK"

]

inventory_summary["inventory_status"] = np.select(

    conditions,

    choices,

    default="OPTIMAL"

)

# ============================================================
# STOCKOUT RISK SCORE
# ============================================================

inventory_summary["stockout_risk_score"] = (

    inventory_summary["reorder_point"]

    /

    np.maximum(
        inventory_summary["current_stock"],
        1
    )

)

inventory_summary["stockout_risk_score"] = (

    inventory_summary["stockout_risk_score"]

    .round(2)

)

# ============================================================
# INVENTORY RECOMMENDATION
# ============================================================

inventory_summary["recommended_order_qty"] = np.where(

    inventory_summary["inventory_status"]
    ==
    "REORDER_REQUIRED",

    np.maximum(

        inventory_summary["eoq"],

        inventory_summary["reorder_point"]

        -

        inventory_summary["current_stock"]

    ),

    0

)

inventory_summary["recommended_order_qty"] = (

    inventory_summary["recommended_order_qty"]

    .round(0)

)

# ============================================================
# INVENTORY VALUE
# ============================================================

inventory_summary["inventory_value"] = (

    inventory_summary["current_stock"]

    *

    inventory_summary["unit_cost"]

)

inventory_summary["recommended_order_value"] = (

    inventory_summary["recommended_order_qty"]

    *

    inventory_summary["unit_cost"]

)

# ============================================================
# ABC ANALYSIS
# ============================================================

inventory_summary["annual_sales_value"] = (

    inventory_summary["annual_demand"]

    *

    inventory_summary["unit_cost"]

)

inventory_summary = (

    inventory_summary

    .sort_values(
        "annual_sales_value",
        ascending=False
    )

)

inventory_summary["cum_sales_pct"] = (

    inventory_summary["annual_sales_value"]

    .cumsum()

    /

    inventory_summary["annual_sales_value"].sum()

)

inventory_summary["abc_class"] = np.where(

    inventory_summary["cum_sales_pct"] <= 0.80,

    "A",

    np.where(

        inventory_summary["cum_sales_pct"] <= 0.95,

        "B",

        "C"

    )

)

# ============================================================
# STOCKOUT REPORT
# ============================================================

stockout_report = (

    inventory_summary

    .query(
        "inventory_status == 'REORDER_REQUIRED'"
    )

    .sort_values(
        "stockout_risk_score",
        ascending=False
    )

)

# ============================================================
# OVERSTOCK REPORT
# ============================================================

overstock_report = (

    inventory_summary

    .query(
        "inventory_status == 'OVERSTOCK'"
    )

    .sort_values(
        "inventory_value",
        ascending=False
    )

)

# ============================================================
# INVENTORY KPI DASHBOARD
# ============================================================

inventory_kpis = pd.DataFrame({

    "Metric":[

        "Total Product-Store Combinations",

        "Total Inventory Value",

        "Products Requiring Reorder",

        "Overstocked Products",

        "Optimal Inventory Products",

        "Total Recommended Order Value",

        "Average Safety Stock",

        "Average Reorder Point"

    ],

    "Value":[

        len(inventory_summary),

        round(
            inventory_summary["inventory_value"].sum(),
            2
        ),

        (
            inventory_summary["inventory_status"]
            ==
            "REORDER_REQUIRED"
        ).sum(),

        (
            inventory_summary["inventory_status"]
            ==
            "OVERSTOCK"
        ).sum(),

        (
            inventory_summary["inventory_status"]
            ==
            "OPTIMAL"
        ).sum(),

        round(
            inventory_summary[
                "recommended_order_value"
            ].sum(),
            2
        ),

        round(
            inventory_summary[
                "safety_stock"
            ].mean(),
            2
        ),

        round(
            inventory_summary[
                "reorder_point"
            ].mean(),
            2
        )

    ]

})

print("\nINVENTORY KPI DASHBOARD")
display(inventory_kpis)

# ============================================================
# TOP STOCKOUT RISKS
# ============================================================

print("\nTOP 20 STOCKOUT RISKS")

display(

    stockout_report[

        [

            "product_id",

            "store_id",

            "current_stock",

            "reorder_point",

            "stockout_risk_score",

            "recommended_order_qty"

        ]

    ]

    .head(20)

)

# ============================================================
# TOP OVERSTOCK RISKS
# ============================================================

print("\nTOP 20 OVERSTOCK PRODUCTS")

display(

    overstock_report[

        [

            "product_id",

            "store_id",

            "current_stock",

            "inventory_value",

            "inventory_status"

        ]

    ]

    .head(20)

)

# ============================================================
# INVENTORY OPTIMIZATION SUMMARY
# ============================================================

inventory_summary_report = pd.DataFrame({

    "Metric":[

        "Service Level",

        "Average Forecast Demand",

        "Average Safety Stock",

        "Average EOQ",

        "Products To Reorder",

        "Overstock Products",

        "Inventory Value"

    ],

    "Value":[

        SERVICE_LEVEL,

        round(
            inventory_summary[
                "avg_daily_forecast"
            ].mean(),
            2
        ),

        round(
            inventory_summary[
                "safety_stock"
            ].mean(),
            2
        ),

        round(
            inventory_summary[
                "eoq"
            ].mean(),
            2
        ),

        len(stockout_report),

        len(overstock_report),

        round(
            inventory_summary[
                "inventory_value"
            ].sum(),
            2
        )

    ]

})

print("\nINVENTORY OPTIMIZATION SUMMARY")
display(inventory_summary_report)

# ============================================================
# EXPORT FOR STEP 15
# ============================================================

inventory_artifacts = {

    "inventory_summary":
    inventory_summary,

    "stockout_report":
    stockout_report,

    "overstock_report":
    overstock_report,

    "inventory_kpis":
    inventory_kpis,

    "service_level":
    SERVICE_LEVEL

}

print("\nStep 14 Completed Successfully")
print("="*120)
````

## 18.HYPERPARAMETER OPTIMIZATION (OPTUNA + MLFLOW)

````python
# ============================================================
# STEP 18 : HYPERPARAMETER OPTIMIZATION (OPTUNA + MLFLOW)
# ============================================================
# Objectives:
# 1. Optimize Forecasting Model Performance
# 2. Tune Hyperparameters Using Optuna
# 3. Use Time-Series Cross Validation
# 4. Track Every Trial with MLflow
# 5. Compare Against Baseline Model
# 6. Save Best Hyperparameters
# 7. Train Final Optimized Model
# 8. Evaluate Optimized Model
# 9. Register Optimized Model
# 10. Export Optimization Artifacts
# ============================================================

import warnings
warnings.filterwarnings("ignore")

import json
import optuna
import mlflow
import mlflow.sklearn

import numpy as np
import pandas as pd

from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import TimeSeriesSplit
from sklearn.metrics import (
    mean_absolute_error,
    mean_squared_error,
    r2_score
)

print("="*120)
print("STEP 18 : HYPERPARAMETER OPTIMIZATION")
print("="*120)

# ============================================================
# LOAD DATA FROM PREVIOUS STEPS
# ============================================================

X_train = forecast_datasets["X_train"]
X_test = forecast_datasets["X_test"]

y_train = forecast_datasets["y_train"]
y_test = forecast_datasets["y_test"]

baseline_model = model_artifacts["best_model"]

baseline_name = model_artifacts["best_model_name"]

# ============================================================
# BASELINE PERFORMANCE
# ============================================================

baseline_pred = baseline_model.predict(X_test)

baseline_mae = mean_absolute_error(
    y_test,
    baseline_pred
)

baseline_rmse = np.sqrt(
    mean_squared_error(
        y_test,
        baseline_pred
    )
)

baseline_r2 = r2_score(
    y_test,
    baseline_pred
)

print("\nBaseline Performance")
print(f"Model : {baseline_name}")
print(f"MAE   : {baseline_mae:.4f}")
print(f"RMSE  : {baseline_rmse:.4f}")
print(f"R2    : {baseline_r2:.4f}")

# ============================================================
# MLFLOW EXPERIMENT
# ============================================================

mlflow.set_experiment(
    "Demand_Forecasting_HPO"
)

# ============================================================
# TIME SERIES CROSS VALIDATION
# ============================================================

tscv = TimeSeriesSplit(
    n_splits=5
)

# ============================================================
# OPTUNA OBJECTIVE FUNCTION
# ============================================================

def objective(trial):

    params = {

        "n_estimators":
        trial.suggest_int(
            "n_estimators",
            100,
            800
        ),

        "max_depth":
        trial.suggest_int(
            "max_depth",
            5,
            30
        ),

        "min_samples_split":
        trial.suggest_int(
            "min_samples_split",
            2,
            20
        ),

        "min_samples_leaf":
        trial.suggest_int(
            "min_samples_leaf",
            1,
            10
        ),

        "max_features":
        trial.suggest_categorical(
            "max_features",
            [
                "sqrt",
                "log2",
                None
            ]
        ),

        "random_state":
        42,

        "n_jobs":
        -1
    }

    cv_scores = []

    with mlflow.start_run(
        nested=True,
        run_name=f"trial_{trial.number}"
    ):

        mlflow.log_params(params)

        for fold, (train_idx, valid_idx) in enumerate(
            tscv.split(X_train)
        ):

            X_tr = X_train.iloc[train_idx]
            X_val = X_train.iloc[valid_idx]

            y_tr = y_train.iloc[train_idx]
            y_val = y_train.iloc[valid_idx]

            model = RandomForestRegressor(
                **params
            )

            model.fit(
                X_tr,
                y_tr
            )

            pred = model.predict(
                X_val
            )

            rmse = np.sqrt(
                mean_squared_error(
                    y_val,
                    pred
                )
            )

            cv_scores.append(rmse)

        mean_rmse = np.mean(
            cv_scores
        )

        mlflow.log_metric(
            "cv_rmse",
            mean_rmse
        )

    return mean_rmse

# ============================================================
# RUN OPTUNA
# ============================================================

N_TRIALS = 30

print("\nStarting Hyperparameter Search...")
print(f"Trials : {N_TRIALS}")

study = optuna.create_study(
    direction="minimize",
    study_name="DemandForecastOptimization"
)

study.optimize(
    objective,
    n_trials=N_TRIALS,
    show_progress_bar=True
)

# ============================================================
# BEST PARAMETERS
# ============================================================

best_params = study.best_params

print("\nBest Parameters")
print(json.dumps(
    best_params,
    indent=4
))

# ============================================================
# TRAIN FINAL OPTIMIZED MODEL
# ============================================================

print("\nTraining Optimized Model...")

optimized_model = RandomForestRegressor(
    **best_params,
    random_state=42,
    n_jobs=-1
)

optimized_model.fit(
    X_train,
    y_train
)

# ============================================================
# FINAL EVALUATION
# ============================================================

optimized_pred = optimized_model.predict(
    X_test
)

optimized_mae = mean_absolute_error(
    y_test,
    optimized_pred
)

optimized_rmse = np.sqrt(
    mean_squared_error(
        y_test,
        optimized_pred
    )
)

optimized_r2 = r2_score(
    y_test,
    optimized_pred
)

optimized_mape = (
    np.mean(
        np.abs(
            (y_test - optimized_pred)
            /
            np.maximum(y_test,1)
        )
    )
    * 100
)

# ============================================================
# IMPROVEMENT ANALYSIS
# ============================================================

rmse_improvement = (
    (
        baseline_rmse
        -
        optimized_rmse
    )
    /
    baseline_rmse
) * 100

mae_improvement = (
    (
        baseline_mae
        -
        optimized_mae
    )
    /
    baseline_mae
) * 100

# ============================================================
# LOG BEST MODEL TO MLFLOW
# ============================================================

with mlflow.start_run(
    run_name="Best_Optimized_Model"
):

    mlflow.log_params(
        best_params
    )

    mlflow.log_metric(
        "mae",
        optimized_mae
    )

    mlflow.log_metric(
        "rmse",
        optimized_rmse
    )

    mlflow.log_metric(
        "r2",
        optimized_r2
    )

    mlflow.log_metric(
        "mape",
        optimized_mape
    )

    mlflow.log_metric(
        "rmse_improvement_pct",
        rmse_improvement
    )

    mlflow.log_metric(
        "mae_improvement_pct",
        mae_improvement
    )

    mlflow.sklearn.log_model(
        optimized_model,
        artifact_path="optimized_model"
    )

# ============================================================
# FEATURE IMPORTANCE
# ============================================================

feature_importance = pd.DataFrame({

    "feature":
    X_train.columns,

    "importance":
    optimized_model.feature_importances_

})

feature_importance = (

    feature_importance

    .sort_values(
        "importance",
        ascending=False
    )

)

# ============================================================
# OPTIMIZATION REPORT
# ============================================================

optimization_report = pd.DataFrame({

    "Metric":[
        "Baseline MAE",
        "Optimized MAE",
        "Baseline RMSE",
        "Optimized RMSE",
        "Baseline R2",
        "Optimized R2",
        "MAPE",
        "RMSE Improvement %",
        "MAE Improvement %"
    ],

    "Value":[
        baseline_mae,
        optimized_mae,
        baseline_rmse,
        optimized_rmse,
        baseline_r2,
        optimized_r2,
        optimized_mape,
        rmse_improvement,
        mae_improvement
    ]

})

print("\nOPTIMIZATION RESULTS")
display(optimization_report)

print("\nTOP 20 FEATURES")
display(
    feature_importance.head(20)
)

# ============================================================
# SAVE ARTIFACTS
# ============================================================

feature_importance.to_csv(
    "optimized_feature_importance.csv",
    index=False
)

optimization_report.to_csv(
    "optimization_report.csv",
    index=False
)

with open(
    "best_hyperparameters.json",
    "w"
) as f:

    json.dump(
        best_params,
        f,
        indent=4
    )

# ============================================================
# EXPORT FOR STEP 19
# ============================================================

optimization_artifacts = {

    "optimized_model":
    optimized_model,

    "best_params":
    best_params,

    "optimization_report":
    optimization_report,

    "feature_importance":
    feature_importance,

    "optimized_predictions":
    optimized_pred,

    "optimized_mae":
    optimized_mae,

    "optimized_rmse":
    optimized_rmse,

    "optimized_r2":
    optimized_r2,

    "optimized_mape":
    optimized_mape
}

print("\nHyperparameter Optimization Completed Successfully")

print(f"\nBest RMSE : {optimized_rmse:.4f}")
print(f"Best MAE  : {optimized_mae:.4f}")
print(f"Best R²   : {optimized_r2:.4f}")

print("="*120)
````

## 20.MODEL EXPLAINABILITY & INTERPRETABILITY (SHAP)

````python
# ============================================================
# STEP 20 : MODEL EXPLAINABILITY & INTERPRETABILITY (SHAP)
# ============================================================
# Objectives:
# 1. Global Feature Importance Analysis
# 2. SHAP Explainability
# 3. Feature Contribution Analysis
# 4. Local Prediction Explanations
# 5. Business Driver Identification
# 6. Demand Driver Ranking
# 7. Explainability Report Generation
# 8. Model Trust & Transparency Assessment
# 9. Save Explainability Artifacts
# 10. Production Explainability Package
# ============================================================

import warnings
warnings.filterwarnings("ignore")

import os
import json
import shap
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.inspection import permutation_importance

print("=" * 120)
print("STEP 20 : MODEL EXPLAINABILITY & INTERPRETABILITY")
print("=" * 120)

# ============================================================
# LOAD ARTIFACTS
# ============================================================

model = optimization_artifacts["optimized_model"]

X_train = forecast_datasets["X_train"]
X_test = forecast_datasets["X_test"]

y_test = forecast_datasets["y_test"]

feature_names = list(X_train.columns)

# ============================================================
# CREATE OUTPUT DIRECTORY
# ============================================================

os.makedirs(
    "artifacts/explainability",
    exist_ok=True
)

# ============================================================
# SAMPLE DATA FOR SHAP
# ============================================================

SHAP_SAMPLE_SIZE = min(
    5000,
    len(X_test)
)

X_sample = X_test.sample(
    SHAP_SAMPLE_SIZE,
    random_state=42
)

print(f"\nSHAP Sample Size: {len(X_sample):,}")

# ============================================================
# GLOBAL FEATURE IMPORTANCE
# ============================================================

print("\nComputing Model Feature Importance...")

if hasattr(model, "feature_importances_"):

    model_importance_df = pd.DataFrame({

        "feature": feature_names,

        "importance":
        model.feature_importances_

    })

else:

    model_importance_df = pd.DataFrame({

        "feature": feature_names,

        "importance":
        np.zeros(len(feature_names))

    })

model_importance_df = (
    model_importance_df
    .sort_values(
        by="importance",
        ascending=False
    )
    .reset_index(drop=True)
)

# ============================================================
# PERMUTATION IMPORTANCE
# ============================================================

print("Computing Permutation Importance...")

perm_importance = permutation_importance(

    model,

    X_sample,

    y_test.loc[X_sample.index],

    n_repeats=5,

    random_state=42,

    scoring="neg_root_mean_squared_error"

)

permutation_df = pd.DataFrame({

    "feature": feature_names,

    "permutation_importance":
    perm_importance.importances_mean

})

permutation_df = (
    permutation_df
    .sort_values(
        by="permutation_importance",
        ascending=False
    )
)

# ============================================================
# SHAP EXPLAINER
# ============================================================

print("Building SHAP Explainer...")

try:

    explainer = shap.TreeExplainer(model)

    shap_values = explainer.shap_values(
        X_sample
    )

except Exception:

    explainer = shap.Explainer(
        model.predict,
        X_sample
    )

    shap_values = explainer(
        X_sample
    )

# ============================================================
# SHAP IMPORTANCE
# ============================================================

print("Computing SHAP Feature Importance...")

if isinstance(shap_values, np.ndarray):

    mean_abs_shap = np.abs(
        shap_values
    ).mean(axis=0)

elif hasattr(shap_values, "values"):

    mean_abs_shap = np.abs(
        shap_values.values
    ).mean(axis=0)

else:

    mean_abs_shap = np.zeros(
        len(feature_names)
    )

shap_importance_df = pd.DataFrame({

    "feature": feature_names,

    "mean_abs_shap":
    mean_abs_shap

})

shap_importance_df = (
    shap_importance_df
    .sort_values(
        by="mean_abs_shap",
        ascending=False
    )
    .reset_index(drop=True)
)

# ============================================================
# DEMAND DRIVER ANALYSIS
# ============================================================

top_drivers = (
    shap_importance_df
    .head(15)
    .copy()
)

top_drivers["driver_rank"] = (
    range(
        1,
        len(top_drivers)+1
    )
)

# ============================================================
# LOCAL EXPLANATIONS
# ============================================================

print("Generating Local Explanations...")

LOCAL_SAMPLES = 10

local_explanations = []

sample_indices = (
    X_sample
    .sample(
        LOCAL_SAMPLES,
        random_state=42
    )
    .index
)

for idx in sample_indices:

    row = X_sample.loc[[idx]]

    prediction = float(
        model.predict(row)[0]
    )

    if hasattr(shap_values, "values"):

        shap_row = (
            shap_values.values[
                X_sample.index.get_loc(idx)
            ]
        )

    else:

        shap_row = (
            shap_values[
                X_sample.index.get_loc(idx)
            ]
        )

    contribution_df = pd.DataFrame({

        "feature":
        feature_names,

        "contribution":
        shap_row

    })

    contribution_df = (
        contribution_df
        .reindex(
            contribution_df.contribution
            .abs()
            .sort_values(
                ascending=False
            ).index
        )
    )

    local_explanations.append({

        "sample_id":
        int(idx),

        "prediction":
        prediction,

        "top_positive_drivers":
        contribution_df
        .head(5)
        .to_dict(
            orient="records"
        )

    })

# ============================================================
# SHAP SUMMARY PLOT
# ============================================================

print("Creating SHAP Visualizations...")

try:

    plt.figure(figsize=(12,8))

    if hasattr(shap_values, "values"):

        shap.summary_plot(

            shap_values.values,

            X_sample,

            show=False

        )

    else:

        shap.summary_plot(

            shap_values,

            X_sample,

            show=False

        )

    plt.tight_layout()

    plt.savefig(

        "artifacts/explainability/shap_summary.png",

        bbox_inches="tight"

    )

    plt.close()

except Exception as e:

    print(
        f"SHAP Summary Plot Error: {e}"
    )

# ============================================================
# FEATURE IMPORTANCE VISUALIZATION
# ============================================================

plt.figure(figsize=(10,8))

top20 = (
    shap_importance_df
    .head(20)
    .sort_values(
        "mean_abs_shap"
    )
)

plt.barh(

    top20["feature"],

    top20["mean_abs_shap"]

)

plt.title(
    "Top 20 Demand Drivers (SHAP)"
)

plt.xlabel(
    "Mean Absolute SHAP Value"
)

plt.tight_layout()

plt.savefig(

    "artifacts/explainability/top20_shap_features.png"

)

plt.close()

# ============================================================
# BUSINESS INTERPRETABILITY REPORT
# ============================================================

business_report = {

    "model_type":
    type(model).__name__,

    "feature_count":
    len(feature_names),

    "top_10_demand_drivers":
    shap_importance_df
    .head(10)
    .to_dict(
        orient="records"
    ),

    "strongest_driver":
    shap_importance_df.iloc[0]["feature"],

    "explainability_method":
    "SHAP",

    "local_explanations":
    LOCAL_SAMPLES

}

# ============================================================
# TRUST SCORE
# ============================================================

top5_contribution = (

    shap_importance_df
    .head(5)["mean_abs_shap"]
    .sum()

)

total_contribution = (
    shap_importance_df[
        "mean_abs_shap"
    ].sum()
)

trust_score = round(

    (
        top5_contribution
        /
        total_contribution
    ) * 100,

    2

)

# ============================================================
# EXPLAINABILITY SUMMARY
# ============================================================

explainability_summary = pd.DataFrame({

    "Metric":[

        "Total Features",

        "SHAP Samples",

        "Top Driver",

        "Top Driver Impact",

        "Trust Score (%)",

        "Local Explanations"

    ],

    "Value":[

        len(feature_names),

        SHAP_SAMPLE_SIZE,

        shap_importance_df.iloc[0]["feature"],

        round(
            shap_importance_df.iloc[0]["mean_abs_shap"],
            4
        ),

        trust_score,

        LOCAL_SAMPLES

    ]

})

# ============================================================
# SAVE ARTIFACTS
# ============================================================

model_importance_df.to_csv(

    "artifacts/explainability/model_feature_importance.csv",

    index=False

)

permutation_df.to_csv(

    "artifacts/explainability/permutation_importance.csv",

    index=False

)

shap_importance_df.to_csv(

    "artifacts/explainability/shap_importance.csv",

    index=False

)

with open(

    "artifacts/explainability/local_explanations.json",

    "w"

) as f:

    json.dump(

        local_explanations,

        f,

        indent=4,

        default=str

    )

with open(

    "artifacts/explainability/business_interpretability_report.json",

    "w"

) as f:

    json.dump(

        business_report,

        f,

        indent=4

    )

# ============================================================
# DISPLAY RESULTS
# ============================================================

print("\nTOP 20 DEMAND DRIVERS")
display(
    shap_importance_df.head(20)
)

print("\nEXPLAINABILITY SUMMARY")
display(
    explainability_summary
)

# ============================================================
# EXPORT FOR STEP 21
# ============================================================

explainability_artifacts = {

    "explainer":
    explainer,

    "shap_importance":
    shap_importance_df,

    "permutation_importance":
    permutation_df,

    "local_explanations":
    local_explanations,

    "business_report":
    business_report,

    "trust_score":
    trust_score,

    "summary":
    explainability_summary

}

print("\nGenerated Artifacts")
print("-" * 50)
print("✓ shap_summary.png")
print("✓ top20_shap_features.png")
print("✓ shap_importance.csv")
print("✓ permutation_importance.csv")
print("✓ local_explanations.json")
print("✓ business_interpretability_report.json")

print("\nStep 20 Completed Successfully")
print("=" * 120)
````
